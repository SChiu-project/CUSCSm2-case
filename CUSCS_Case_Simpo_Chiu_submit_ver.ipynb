### 4. Data processing

```python
#combine 2 data set
titanic_combined = [train_data, eval_data]
```

```python
# label : "survived"
for dataset in titanic_combined:
  print(pd.value_counts(dataset.survived))
```

```python
# label : "sex"
for dataset in titanic_combined:
  print(pd.value_counts(dataset.sex))
```

```python
print (train_data[['sex', 'survived']].groupby(['sex'], as_index=False).mean())
```

```python
# convert the sex variables into numbers
for dataset in titanic_combined:
    dataset['sex'] = dataset['sex'].map( {'female': 1, 'male': 0} ).astype(int)
```

```python
print (train_data[['sex', 'survived']].groupby(['sex'], as_index=False).mean())
```

```python
# label : "age"
for dataset in titanic_combined:
  print(pd.value_counts(dataset.age))
```

```python
# create 5 groups for ages
train_data['AgeBand'] = pd.cut(train_data['age'], 5)

print (train_data[['AgeBand', 'survived']].groupby(['AgeBand'], as_index=False).mean())
```

```python
# map Age according to AgeBand
for dataset in titanic_combined:
    dataset.loc[ dataset['age'] <= 16.6, 'age'] = 0
    dataset.loc[(dataset['age'] > 16.6) & (dataset['age'] <= 32.5), 'age'] = 1
    dataset.loc[(dataset['age'] > 32.5) & (dataset['age'] <= 48.3), 'age'] = 2
    dataset.loc[(dataset['age'] > 48.3) & (dataset['age'] <= 64.2), 'age'] = 3
    dataset.loc[ dataset['age'] > 64.2, 'age'] = 4
    dataset['age'] = dataset['age'].astype(int)
```

```python
# double check the distribution
print (train_data[['age', 'survived']].groupby(['age'], as_index=False).mean())
```

```python
one_hot_train_data = pd.get_dummies(train_data['age'])

train_data = pd.concat([train_data, one_hot_train_data], axis=1)

titanic_combined[0] = train_data
```

```python
one_hot_eval_data = pd.get_dummies(eval_data['age'])

eval_data = pd.concat([eval_data, one_hot_eval_data], axis=1)

titanic_combined[1] = eval_data
```

```python
# label : "n_siblings_spouses"
for dataset in titanic_combined:
  print(pd.value_counts(dataset.n_siblings_spouses))
```

```python
# label : "parch"
for dataset in titanic_combined:
  print(pd.value_counts(dataset.parch))
```

```python
# new variable "FamilySize" to combine "n_siblings_spouses" and "parch"
for dataset in titanic_combined:
    dataset['FamilySize'] = dataset['n_siblings_spouses'] +  dataset['parch'] + 1
    print (dataset[['FamilySize', 'survived']].groupby(['FamilySize'], as_index=False).mean())
```

```python
train_data
```

```python
# map into different groups
for dataset in titanic_combined:
    dataset.loc[ dataset['FamilySize'] <= 1, 'FamilySize'] = 0
    dataset.loc[(dataset['FamilySize'] > 1) & (dataset['FamilySize'] <= 4), 'FamilySize'] = 1
    dataset.loc[ dataset['FamilySize'] > 4, 'FamilySize'] = 2
    dataset['FamilySize'] = dataset['FamilySize'].astype(int)
```

```python
# double check distribution
print (train_data[['FamilySize', 'survived']].groupby(['FamilySize'], as_index=False).mean())
```

```python
one_hot_train_data = pd.get_dummies(train_data['FamilySize'])

train_data = pd.concat([train_data, one_hot_train_data], axis=1)

titanic_combined[0] = train_data
```

```python
one_hot_eval_data = pd.get_dummies(eval_data['FamilySize'])

eval_data = pd.concat([eval_data, one_hot_eval_data], axis=1)

titanic_combined[1] = eval_data
```

```python
one_hot_eval_data = pd.get_dummies(eval_data['fare'])

eval_data = pd.concat([eval_data, one_hot_eval_data], axis=1)

titanic_combined[1] = eval_data
```

### 5. ML Model 1 - Decision Tree


```python
#training data
X_train=train_data.drop('survived',axis=1)
y_train=train_data['survived']
```

```python
from sklearn.tree import DecisionTreeClassifier

dt_clf = DecisionTreeClassifier()
dt_clf.fit(X_train.values, y_train.values)
```

### 6. ML Model 2 - SVM

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC

clf = make_pipeline(StandardScaler(), SVC(gamma='auto'))
clf.fit(X_train.values, y_train.values)
```

### 7. Evaluation

```python
X_eval=eval_data.drop('survived',axis=1)
y_eval=eval_data['survived']
```

```python
from sklearn.metrics import classification_report,confusion_matrix

# for ML Model 1 - Decision tree
y_predict1 = dt_clf.predict(X_eval.values)
print(classification_report(y_eval,y_predict1))
```

```python
# for ML Model 2 - SVM
y_eval_predict2 = clf.predict(X_eval.values)
print(classification_report(y_eval,y_eval_predict2))
```

### 8. Recommendation

According to the result in confusion matrix tested with the evaluation data, Decision Tree perform better than SVM model in Accuracy score
